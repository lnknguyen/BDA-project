```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import warnings

from itertools import islice
from community import community_louvain
from sklearn.preprocessing import normalize
from scipy.stats import pearsonr
from mpl_toolkits.axes_grid1 import make_axes_locatable

warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame(pd.read_csv('train.csv'))
```

```python
unique_neighbor = sorted(df['Neighborhood'].unique())
all_neighbor = df["Neighborhood"].to_list()
neighbor_as_numeric_value = [unique_neighbor.index(neigh) for neigh in all_neighbor]
```

```python
fig = plt.figure()
fig.set_figwidth(25)
fig.set_figheight(15)

ax = fig.add_subplot(111, projection = '3d')

df["TotalArea"] = df["GrLivArea"] + df["TotalBsmtSF"]


img = ax.scatter(neighbor_as_numeric_value,
                 df["OverallQual"],
                 df["TotalArea"],
                 c = df["SalePrice"], 
                 cmap = plt.cm.rainbow)

plt.xticks(neighbor_as_numeric_value, all_neighbor)
plt.xticks(rotation = 45)

fig.colorbar(img,fraction = 0.02, pad = 0.04)
plt.show()
```
